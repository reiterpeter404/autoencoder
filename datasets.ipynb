# Tensorflow Keras Datasets

```python
import keras.datasets as data
from matplotlib import pyplot as plt
```

```python
def visualize_images(_data, _rows = 2, _columns = 8):
    fig, axes = plt.subplots(_rows, _columns, figsize = (12, 2))

    for i in range(_rows):
        for j in range(_columns):
            # calculate the index of the current element and print the picture
            cnt = i * _columns + j
            axes[i,j].imshow(_data[cnt], cmap= 'gray')
```
